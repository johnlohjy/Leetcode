```python
class Solution(object):
    def openLock(self, deadends, target):
        """
        :type deadends: List[str]
        :type target: str
        :rtype: int
        """

        """
        Lock starts at 0000

        deadends: If lock displays any of these, wheels will stop turning

        Target: unlock

        Return the min total number of turns

        Hint:
        We can think of this problem as a shortest path problem on a graph: there are `10000` nodes (strings `'0000'` to `'9999'`), and there is an edge between two nodes if they differ in one digit, that digit differs by 1 (wrapping around, so `'0'` and `'9'` differ by 1), and if *both* nodes are not in `deadends`.
        
        
        0000
        
        1000
        2000
        3000
        4000
        5000
        6000
        7000
        8000
        9000

        ##############################

        0100

        0000
        0200
        0300
        0400
        0500
        0600
        0700
        0800
        0900

        #############################
        """

 

        visited = set()
        queue = ["0000"]
        res = 0
        while len(queue) > 0:
            for i in range(len(queue)):
                node = queue.pop(0)
                if node in target:
                    return res
                if node not in visited:
                    visited.add(node)
                    if node not in deadends:
                        for i in range(4):
                            temp = list(node)
                            change = int(node[i])
                            if change==0:
                                change_up = "1"
                                change_down = "9"
                                temp[i]=change_up
                                queue.append(''.join(temp))
                                temp[i]=change_down
                                queue.append(''.join(temp))
                            elif change==9:
                                change_up = "0"
                                change_down = "8"
                                temp[i]=change_up
                                queue.append(''.join(temp))
                                temp[i]=change_down
                                queue.append(''.join(temp))
                            else:
                                change_up = str(change+1)
                                change_down = str(change-1)
                                temp[i]=change_up
                                queue.append(''.join(temp))
                                temp[i]=change_down
                                queue.append(''.join(temp))
            res+=1
        return -1
                            


                    

        
```
