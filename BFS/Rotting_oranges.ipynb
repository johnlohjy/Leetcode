```python
class Solution(object):
    def orangesRotting(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        #Initialise a Queue
        queue = []

        #Initialise a list of visited positions
        positions = []

        #Initialise minutes
        minutes = 0

        #Keep track of total number of oranges
        total_oranges = 0

        #Keep track of number of fresh oranges
        fresh_oranges = 0

        #Initialise number of rows
        num_rows = len(grid)

        #Initialise number of cols
        num_cols = len(grid[0])

        #Initialise directions
        directions = [[0,-1], [0,1], [1,0], [-1,0]]

        #Enqueue all rotten oranges to perform BFS
        #and count total number of oranges
        for row in range(num_rows):
            for col in range(num_cols):
                if(grid[row][col]==2):
                    queue.append((row,col))
                if(grid[row][col]==1):
                    fresh_oranges+=1

                total_oranges+=1

        if(fresh_oranges==0):
            return 0

        #Perform BFS: Need a queue to append the adjacent neighbours
        while(len(queue)>0 and fresh_oranges>0):
            for i in range(len(queue)):
                curr_pos = queue.pop(0)
                row, col = curr_pos

                for d in directions:
                    dr = d[0]
                    dc = d[1]
                    if((row+dr,col+dc) in positions or row+dr<0 or row+dr==num_rows or col+dc<0 or col+dc==num_cols or grid[row+dr][col+dc]!=1):
                        continue
                        
                    positions.append((row+dr,col+dc))
                    queue.append((row+dr,col+dc))
                    grid[row+dr][col+dc] = 2
                    fresh_oranges-=1

            minutes+=1

        if(fresh_oranges==0):
            return minutes
        else:
            return -1
            

            
```
