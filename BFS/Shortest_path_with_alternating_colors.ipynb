```python
class Solution(object):
    def shortestAlternatingPaths(self, n, redEdges, blueEdges):
        """
        :type n: int
        :type redEdges: List[List[int]]
        :type blueEdges: List[List[int]]
        :rtype: List[int]
        """

        """
        Hint: 
        Do a breadth-first search, where the "nodes" are actually (Node, color of last edge taken).

        n nodes
        labelled 0 to n-1
        Each edge: red or blue 
        - self edge
        - parallel edge

        redEdges[i] = [a,b]
        directed red edge from node a to node b

        blueEdges[i] = [u,v]
        directed blue edge from node u to node v

        return arr answer of length n
        answer[x]: length of shortest path from node 0 to node x s.t edge colours alternate along the path

        -1 if such a path dne

        Example: 
        n = 3, redEdges = [[0,1],[1,2]], blueEdges = []
        Graph
        0  -red->  1  -red->  2

        Example:
        n = 3, redEdges = [[0,1]], blueEdges = [[2,1]]
        Graph
        0  -red->  1  <-blue- 2

        Own Example:
        0  -red->  1  <-blue- , -red->  2

        Red
        {
            0: 1,
            1: 2
        }

        Blue
        {
            2: 1
        }
        """

        adj_list_red = defaultdict(list)
        adj_list_blue = defaultdict(list)
        answer_red = [10000] * n
        answer_blue = [10000] * n

        for edge in redEdges:
            adj_list_red[edge[0]].append(edge[1])

        for edge in blueEdges:
            adj_list_blue[edge[0]].append(edge[1])

        queue_red = [0]
        queue_blue = [0]
        visited_red = defaultdict(int)
        visited_blue = defaultdict(int)
        time_red = 0
        time_blue = 0
        while len(queue_red) > 0:
            for i in range(len(queue_red)):
                node = queue_red.pop(0)

                if visited_red[node] < 100:
                    answer_red[node] = min(answer_red[node],time_red)
                    visited_red[node]+=1

                    if time_red % 2 == 0:
                        for nb in adj_list_red[node]:
                            queue_red.append(nb)

                    else:
                        for nb in adj_list_blue[node]:
                            queue_red.append(nb)
            time_red+=1

        while len(queue_blue) > 0:
            for i in range(len(queue_blue)):
                node = queue_blue.pop(0)

                if visited_blue[node] < 100:
                    answer_blue[node] = min(answer_blue[node],time_blue)
                    visited_blue[node]+=1

                    if time_blue % 2 == 0:
                        for nb in adj_list_blue[node]:
                            queue_blue.append(nb)

                    else:
                        for nb in adj_list_red[node]:
                            queue_blue.append(nb)
            time_blue+=1

        answer = [-1]*n


        for i in range(n):
            answer[i] = min(answer_blue[i],answer_red[i])
            if answer[i]==10000:
                answer[i]=-1

        return answer


        


        

                

            




               

                





        
```
