```python
# KIV
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    lca = None
    deepest_lca = 0
    def lcaDeepestLeaves(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        
        """
        Do a postorder traversal

        Then, if both subtrees contain a deepest leaf, you can mark this node as the answer (so far).

        The final node marked will be the correct answer.

        Binary Tree Traversals:

        1) Post order traversal:
        Left-Right-Root Policy
        The left subtree is traversed first
        Then the right subtree is traversed
        Finally, the root node of the subtree is traversed

        pseudocode:
        def postorder(node):
            if node == None:
                return
            postorder(node.left)
            postorder(node.right)
            print(node.val)

        2) In order traversal:
        Left-Root-Right Policy
        The left subtree is traversed first
        Then the root node for that subtree is traversed
        Finally, the right subtree is traversed

        pseudocode:
        def inorder(node):
            if node == None:
                return
            inorder(node.left)
            print(node.val)
            inorder(node.right)


                    0
                1        3
            null  2
        ^
        if the lowest leave is by itself then return that node
        """
        depth = 0
        def postorder(node, depth):
            depth+=1

            # If the node is a candidate deepest node, return True
            if node.left is None and node.right is None:
                return True
            else:
                # Check if its left and right subtrees have the deepest node
                left_deepest = False
                right_deepest = False
                if node.left:
                    left_deepest = postorder(node.left, depth)
                if node.right:
                    right_deepest = postorder(node.right, depth)

                # If its left and right subtree has a deepest node and it is the new lowest lca
                # update the depth of the deepest lca and save the node as the lowest lca so far 
                if left_deepest and right_deepest and depth > self.deepest_lca:
                    self.deepest_lca = depth
                    self.lca = node

                # Return if its subtree contains any deepest nodes
                return left_deepest or right_deepest
            
        postorder(root, depth)
        return self.lca

```
