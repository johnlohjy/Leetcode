```python
class Solution(object):
    def shortestBridge(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        """
        1. Use DFS to identify islands by adding them to 2 sets

        2. Do a DFS/BFS from all the cells of the island with less cells and store the min
        """
        island_1 = set()
        island_2 = set()
        visited = set()

        def dfs(r,c,grid,islands,visited):
            if (r,c) not in visited:
                visited.add((r,c))

                if r < 0 or r > len(grid)-1 or c < 0 or c > len(grid[0])-1 or grid[r][c]==0:
                    return
                
                islands.add((r,c))
                dfs(r+1,c,grid,islands,visited)
                dfs(r-1,c,grid,islands,visited)
                dfs(r,c-1,grid,islands,visited)
                dfs(r,c+1,grid,islands,visited)
                return
            return

        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if grid[r][c]==1 and len(island_1)==0:
                    dfs(r,c,grid,island_1,visited)
                elif grid[r][c]==1 and len(island_2)==0:
                    dfs(r,c,grid,island_2,visited)

        min_islands = min(len(island_1),len(island_2))
        directions = [(1,0),(-1,0),(0,-1),(0,1)]

        def bfs(r,c,this_island,other_island,steps,visited,directions):
            queue = [(r,c)]
            while len(queue)>0:
                for i in range(len(queue)):
                    node = queue.pop(0)
                    if node not in visited:
                        visited.add(node)
                        if node in other_island:
                            return steps-1
                        
                        for dr in directions:
                            new_r = node[0]+dr[0]
                            new_c = node[1]+dr[1]
                            if (new_r,new_c) not in this_island:
                                queue.append((new_r,new_c))
                steps+=1




        island_1 = list(island_1)
        island_2 = list(island_2)
        res = []

        if len(island_1)==min_islands:
            for island in island_1:
                visited = set()
                steps = bfs(island[0],island[1],island_1,island_2,0,visited,directions)
                if steps is not None:
                    res.append(steps)
        else:
            for island in island_2:
                visited = set()
                steps = bfs(island[0],island[1],island_2,island_1,0,visited,directions)
                if steps is not None:
                    res.append(steps)

        return min(res)


                



        
```
