```python
class Solution(object):
    def minMutation(self, startGene, endGene, bank):
        """
        :type startGene: str
        :type endGene: str
        :type bank: List[str]
        :rtype: int
        """

        """
        gene string: A, C, G, T

        Investigate a mutation from startGene to endGene
        Mutation is defined as single character changed in the gene string

        "AACCGGTT" --> "AACCGGTA" is one mutation

        bank: stores all valid genes

        Return the minimum number of mutations needed to mutate from startGene to endGene. If there is no such a mutation, return -1.

        startGene = "AACCGGTT", endGene = "AAACGGTA"

        USE LEVEL ORDER TRAVERSAL
        ...
        """

        def can_transform(gene,next_gene):
            diff = 0
            for i in range(len(gene)):
                if gene[i]!=next_gene[i]:
                    diff+=1
            if diff==1:
                return True
            return False
        
        if endGene not in bank:
            return -1

        visited = set()
        adj_list = defaultdict(list)
        all_genes = list(set([startGene] + bank + [endGene]))
        for i in range(len(all_genes)):
            for j in range(len(all_genes)):
                if i!=j:
                    if can_transform(all_genes[i],all_genes[j]):
                        adj_list[all_genes[i]].append(all_genes[j])

        print(adj_list)

        q = [startGene]
        depth = 0
        reached = False
        while len(q) > 0:
            for i in range(len(q)):
                gene = q.pop(0)

                if gene==endGene:
                    reached = True
                    return depth

                visited.add(gene)

                for nb in adj_list[gene]:
                    if nb not in visited:
                        q.append(nb)
            depth+=1
            
        

        if reached:
            return depth
        return -1



        

            

            
            


        

        
```
