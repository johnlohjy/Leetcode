```python
class Solution:
    def nearestExit(self, maze, entrance):
        
        seen = set()
        exits = set()
        #Define num rows and columns
        R, C = len(maze), len(maze[0])
        entrance = tuple(entrance)
        
        #Add all exits
        for r in range(R):
            #If its empty, its an exit. This is for left col
            if maze[r][0] == ".":
                exits.add((r, 0))
            #If its empty, its an exit. This is for right col
            if maze[r][C-1] == ".":
                exits.add((r, C-1))
        
        for c in range(C):
            #If its empty, its an exit. This is for top row
            if maze[0][c] == ".":
                exits.add((0, c))
            #If its empty, its an exit. This is for bottom row
            if maze[R-1][c] == ".":
                exits.add((R-1, c))
        
        #If the entrance is in the exit, remove it from the exits because it's not supposed to be counted as an exit
        if entrance in exits:
            exits.remove(entrance)
        
        
        #Initialise the starting point and the number of steps
        q = [(entrance, 0)] # start, distance
        
        #while the q is not empty
        while q:
            #Get the current position and number of steps
            curr_pos , d = q.pop(0)
            
            #If the current position is in exits, return the number of steps 
            if curr_pos in exits:
                return d
            
            #If it has been seen before, continue
            if curr_pos in seen:
                continue
            
            #Add it to seen
            seen.add(curr_pos)
            
            #For all the possible change in directions
            for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                nx = dx + curr_pos[0]
                ny = dy + curr_pos[1]
                
                #If the new position is within bounds and it is empty, add it to the queue
                if 0 <= nx < R and 0 <= ny < C and maze[nx][ny] == ".":
                    q.append(((nx, ny), d + 1))
            
        
        return -1
```
