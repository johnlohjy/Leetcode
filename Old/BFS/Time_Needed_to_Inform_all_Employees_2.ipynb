```python
class Solution(object):
    def numOfMinutes(self, n, headID, manager, informTime):
        """
        :type n: int
        :type headID: int
        :type manager: List[int]
        :type informTime: List[int]
        :rtype: int
        """

        """
        n employees
        Unique ID for each employee from 0 to n-1
        head of the company has id: headID

        each employee has 1 direct manager
        manager[i] is the direct manager of the ith employee
        manager[headID] = -1

        Scenario:
        The head of the company wants to inform all the company employees of an urgent piece of news. He will inform his direct subordinates, and they will inform their subordinates, and so on until all employees know about the urgent news.

        The i-th employee needs informTime[i] minutes to inform all of his direct subordinates (i.e., After informTime[i] minutes, all his direct subordinates can start spreading the news).

        Return the number of minutes needed to inform all the employees about the urgent news.

        Construct an adjacency list 

        Hints:
        1. The company can be represented as a tree, headID is always the root.
        2. Store for each node the time needed to be informed of the news.
        3. Answer is the max time a leaf node needs to be informed.

        Example:

        Input: n = 6, headID = 2, manager = [2,2,-1,2,2,2], informTime = [0,0,1,0,0,0]

        Directed Adj List
        {
            2: [0,1,3,4,5]
        }

        Perform a BFS
        At the curr node, ddd on the time taken in the parent node (time needed to be informed of news so far)
        """

        from collections import defaultdict

        # Initialise a directed adjacency list to represent the tree
        # Remember that tree follows hierarchical structure - big boss to small employee
        directed_adj_list = defaultdict(list)

        # Initialise a dictionary that contains the time it takes for a node to receive the news
        times = defaultdict(int)
        times[headID] = 0

        for i in range(len(manager)):
            # Key: Manager, Value: Children
            directed_adj_list[manager[i]].append(i)

        queue = []
        queue = [headID]

        while len(queue) > 0:
            node = queue.pop(0)
            # For each of the children
            for i in range(len(directed_adj_list[node])):
                # Record the time it takes for the curr child to receive the news as
                # cumulated time its parent took to receive the news + time it takes for its parent to disseminate the news
                times[directed_adj_list[node][i]] = times[node] + informTime[node]

                # Enqueue the children
                queue.append(directed_adj_list[node][i])

        return max(times.values())


        
```
