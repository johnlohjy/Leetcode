```python
class Solution(object):
    def maxDistance(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        """
        Given a nxn grid that contains 0 or 1. 0 = water, 1 = land, find a water cell st its distance to the nearest land
        cell is maximized and return the distance
        Manhatten dist is used: |x0-x1| + |y0-y1|

        Hint:
        Can you think of this problem in a backwards way ?
        Imagine expanding outward from each land cell. What kind of search does that ?
        Use BFS starting from all land cells in the same time.
        When do you reach the furthest water cell?

        Find the cell furthest from all cells 

        Apply multi source BFS
        Water cell that is reached last will be the furthest from all
        """

        q = []
        lands = []
        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if grid[r][c]==1:
                    q.append((r,c))
                    lands.append((r,c))

        if len(lands)==0:
            return -1
        
        visited = set()

        directions = [(1,0),(-1,0),(0,-1),(0,1)]

        water = {}

        time = 0

        while len(q)>0:
            for i in range(len(q)):
                node = q.pop(0)
                if node not in visited:
                    visited.add(node)
                    if grid[node[0]][node[1]]==0:
                        water[(node[0],node[1])] = time
                    for d in directions:
                        if node[0]+d[0] < 0 or node[0]+d[0] > len(grid)-1 or node[1]+d[1] < 0 or node[1]+d[1] > len(grid[0])-1:
                            continue
                        if (node[0]+d[0],node[1]+d[1]) not in visited:
                            q.append((node[0]+d[0],node[1]+d[1]))
            time+=1

        if len(water)==0:
            return -1

        furthest = max(zip(water.values(), water.keys()))[1]

        curr = 1000
        for land in lands:
            man_dist = abs(furthest[0]-land[0]) + abs(furthest[1]-land[1])
            if man_dist < curr:
                curr = man_dist

        return curr

                

                    


        
```
