```python
#NEED TO OPTIMISE

class Solution(object):
    def minReorder(self, n, connections):
        """
        :type n: int
        :type connections: List[List[int]]
        :rtype: int
        """

        """


        Hint:
        Treat the graph as undirected. Start BFS from the root, if you come across an edge in the forward direction, 
        then we need to reverse the edge.

        Initialise an adjacency list where the original edge has a marker of 1 (if the original edge allows us to
        reach the next node in BFS, it is a forward edge and needs to be flipped - indicated by 1), and its opposite edge has a marker of 0
        where 1=>flip, 0=>never flip, forming an undirected graph

        Undirected graph
        - Original edges that are forward egdges are indicated by 1
        - rest of edges are 0

        Perform BFS using an adjacency list where original edges that are forward edges are marked with 1

        """
        
        #Initialise adjacency list
        adj_list = collections.defaultdict(list)
        for u,v in connections:
            adj_list[u].append((v,1))
            adj_list[v].append((u,0))

        #Initialise a queue
        queue = []
        visited = set()

        #Start from root: 0
        queue.append(0)
        visited.add(0)

        #Number of flipped edges
        flips = 0

        #Perform BFS,
        #adding the number of edges to be flipped while doing so: Original edges that are forward edges need to be flipped
        while len(queue) > 0:
            node = queue.pop(0)
            for nb in adj_list[node]:
                #If the neighbour has not been visited before
                if nb[0] not in visited:
                    visited.add(nb[0])
                    queue.append(nb[0])
                    flips+=nb[1]

        return flips
                    
```
