```python
class Solution(object):
    def accountsMerge(self, accounts):
        """
        :type accounts: List[List[str]]
        :rtype: List[List[str]]
        """
        
        """
        Given a list accounts
        accounts[i][0] is a name
        accounts[i][1:] are emails 

        2 accounts belong to the same person if there is a common email to both accounts

        accounts = [
            ["John","johnsmith@mail.com","john_newyork@mail.com"],
            ["John","johnsmith@mail.com","john00@mail.com"],
            ["Mary","mary@mail.com"],
            ["John","johnnybravo@mail.com"]
        ]

        Answer: 
        [
            ["John","john00@mail.com","john_newyork@mail.com","johnsmith@mail.com"],
            ["Mary","mary@mail.com"],
            ["John","johnnybravo@mail.com"]
        ]

        [
            ['Mary', 'mary@mail.com'], 
            ['John', 'johnnybravo@mail.com'], 
            ['John', 'john00@mail.com', 'john_newyork@mail.com', 'johnsmith@mail.com']
        ]

        accounts = [
            ["John","a","b"],
            ["John","a","c"],
            ["Mary","e"],
            ["John","d"]
        ]

        {
            a: b,c
            e:
            d:
        }

        {
            a: b,c
            b: a
            c: a
            e:
            d:
        }



        a b c d e f

        For every pair of emails in the same account, draw an edge between those emails. The problem is about enumerating the connected components of this graph.

        Adapted from: https://leetcode.com/problems/accounts-merge/solutions/2550030/python-construct-graph-and-dfs-easy-implementation/
        """

        adj_list = defaultdict(list)
        for account in accounts:
            for i in range(1,len(account)-1):
                adj_list[account[i]].append(account[i+1])
                adj_list[account[i+1]].append(account[i])

        def dfs(email, adj_list, visited):
            visited.add(email)
            res = [email]
            for nb in adj_list[email]:
                if nb not in visited:
                    res += dfs(nb,adj_list,visited)
            return res

        visited = set()
        result = []
        for email in accounts:
            if email[1] not in visited:
                result.append([email[0]] + sorted(dfs(email[1],adj_list,visited)))
        
        return result
        






        
        
```
