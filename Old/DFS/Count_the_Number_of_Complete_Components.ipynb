```python
class Solution(object):
    def countCompleteComponents(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: int
        """

        """
        n vertices
        labelled 0 to n-1

        edges[i] = [a_i, b_i]

        Connected component: Subgraph where there exists a path between
        any 2 vertices. There exists an edge between every pair
        of its vertices

        Return the number of complete connected components

        Hint:
        Find the connected components of an undirected graph using depth-first search (DFS) or breadth-first search (BFS).

        For each connected component, count the number of nodes and edges in the component.

        A connected component is complete if and only if the number of edges in the component is equal to m*(m-1)/2, where m is the number of nodes in the component.

        Example

        n = 6, edges = [[0,1],[0,2],[1,2],[3,4]]

        {
            0: 1,2
            1: 2
            3: 4
            5
        }

        Find connected components
        Perform dfs using a node
        visit its neighbours recursively and add to a list
        Determine if its a connected component
        - remember to divide number of edges by 2 since we double count
        - think about the case [1,0] [2,0]. Need to traverse all so we include
        edges for both nodes
        using the formula: m*(m-1)/2
        """

        adj_list = defaultdict(list)

        for edge in edges:
            adj_list[edge[0]].append(edge[1])
            adj_list[edge[1]].append(edge[0])

        visited = set()

        def dfs(node,visited,adj_list,connected_components):
            if node not in visited:
                visited.add(node)
                connected_components.append(node)
                for nb in adj_list[node]:
                    dfs(nb,visited,adj_list,connected_components)

        current = 0

        for i in range(n):
            connected_components = []
            dfs(i,visited,adj_list,connected_components)
            num_nodes = len(connected_components)
            edges = 0
            for j in connected_components:
                edges += len(adj_list[j])
            if num_nodes > 0 and num_nodes*(num_nodes-1)/2 == edges/2 :
                current+=1
        
        return current
        
```
