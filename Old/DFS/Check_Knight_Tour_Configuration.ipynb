```python
class Solution(object):
    def checkValidGrid(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: bool
        """

        """
        nxn chessboard
        valid config:
        - knight starts at top left cell
        - visit every cell exactly once

        range of integers: 0 to n*n - 1
        grid[r][c] = grid[r][c]th cell that the knight visited

        valid knight move:
        - move 2 squares vertically and 1 sq horizontally
        - move 2 squares horizontally and 1 sq vertically

        all possible knight moves:
        - grid[r-2][c+1]
        - grid[r-2][c-1]
        - grid[r-1][c+2]
        - grid[r-1][c-2]

        - grid[r+1][c+2]
        - grid[r+1][c-2]
        - grid[r+2][c+1]
        - grid[r+2][c-1]

        Hint:
        It is enough to check if each move of the knight is valid.
        Try all cases of the knight's movements to check if a move is valid.
        """

        visited = set()
        res = []

        def dfs(r,c,grid,visited,expected,res):
            if r < 0 or r > len(grid)-1 or c < 0 or c > len(grid[0])-1 or grid[r][c] != expected:
                return
            
            res.append((r,c))
            next_expected = expected+1
            
            dfs(r-2,c+1,grid,visited,next_expected,res)
            dfs(r-2,c-1,grid,visited,next_expected,res)
            dfs(r-1,c+2,grid,visited,next_expected,res)
            dfs(r-1,c-2,grid,visited,next_expected,res)

            dfs(r+1,c+2,grid,visited,next_expected,res)
            dfs(r+1,c-2,grid,visited,next_expected,res)
            dfs(r+2,c+1,grid,visited,next_expected,res)
            dfs(r+2,c-1,grid,visited,next_expected,res)

            return

        
        if grid[0][0]==0:
            dfs(0,0,grid,visited,0,res)

        if len(res)==len(grid)*len(grid):
            return True
        return False

        """
           0  1  2  3  4
  0      [24,11,22,17,04]
  1      [21,16,05,12,09]
  2      [06,23,10,03,18]
  3      [15,20,01,08,13]
  4      [00,07,14,19,02]
        """
                
                
        
```
