```python
class Solution(object):
    def minScore(self, n, roads):
        """
        :type n: int
        :type roads: List[List[int]]
        :rtype: int
        """

        """
        n: n cities numbered from 1 to n
        roads[i] = [a,b,distance between a and b]
        Score of a path: MINIMUM DIST OF THE ROAD IN THIS PATH
        Return min score of path between city 1 and city n

        Path is a seq of roads between 2 cities
        Path can contain the same road multiple times
        Can revisit 1 and n multiple times

        Just need to find the min score? - use dfs to find cuz some nodes may not be connected

        {
            1: [(2,9), (4,7)],
            2: [(1,9), (4,5), (3,6)],
            3: [(2,6)],
            4: [(1,7), (2,5)]
        }
        """

        adj_list = defaultdict(list)

        for r in roads:
            adj_list[r[0]].append((r[1],r[2]))
            adj_list[r[1]].append((r[0],r[2]))

                

        visited = set()

        all_roads = []
        
        def dfs(node,adj_list,visited,all_roads):
            if node not in visited:
                visited.add(node)
                for nb in adj_list[node]:
                    all_roads.append(nb[1])
                    dfs(nb[0], adj_list, visited, all_roads)
                return
            return
        
        dfs(1,adj_list,visited,all_roads)
        return min(all_roads)




        
```
