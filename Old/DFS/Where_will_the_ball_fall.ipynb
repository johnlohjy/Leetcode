```python
"""
Need to optimise
"""
class Solution(object):
    def findBall(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: List[int]
        """

        """
        Hint: DFS
        Traverse the path of the ball downwards until you reach the bottom or get stuck.

        Cases:
        Condition to go to row+1,col+1 cell:
            current pos and right pos cell are 1
        Condition to go to row+1,col-1 cell:
            current pos and left adj pos are -1
        Stuck:
            all other cases
        """

        #Base Case: 1 cell, ball gets stuck
        if len(grid[0])==1:
            return [-1]

        def dfs(r,c,grid):
            #Stopping condition
            #Fall out cases
            #Check if its within bounds and where it will fall out
            if c+1<=len(grid[0])-1 and r==len(grid)-1 and grid[r][c]==1 and grid[r][c+1]==1:
                return c+1
            elif c-1>=0 and r==len(grid)-1 and grid[r][c]==-1 and grid[r][c-1]==-1:
                return c-1
            #Check if its within bounds and if ball can go bottom right
            elif c+1<=len(grid[0])-1 and grid[r][c]==1 and grid[r][c+1]==1:
                return dfs(r+1,c+1,grid)
            #Check if its within bounds and if ball can go bottom left
            elif c-1>=0 and grid[r][c]==-1 and grid[r][c-1]==-1:
                return dfs(r+1,c-1,grid)
            #Else: Stuck
            else:
                return -1

        res = []

        for c in range(len(grid[0])):
            res.append(dfs(0,c,grid))

        return res
```
