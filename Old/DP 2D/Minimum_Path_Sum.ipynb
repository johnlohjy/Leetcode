```python
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        """
        Find path from top left to bottom right with minimum sum

        Can only move down or right at any point in time

        DP
        1. Bottom up
        What are things we need to build up
        What is the base case
        2. Top Down
        What is the sub-problem/recurrence relation for top down
        What is the base case

        2D DP problem

        Top Down Approach: At each cell, what is the minimum path sum to reach to that cell
        - can only come from up or left

        dp[r][c] = min(dp[r-1][c], dp[r][c-1])

        """


        dp = [[None for _ in range(len(grid[0]))] for _ in range(len(grid))]


        def top_down(r,c,dp,grid):
            if r==0 and c==0:
                return grid[r][c]
            elif r==0:
                return grid[r][c] + top_down(r,c-1,dp,grid)
            elif c==0:
                return grid[r][c] + top_down(r-1,c,dp,grid)
            else:
                if dp[r][c]!=None:
                    return dp[r][c]
                dp[r][c] = grid[r][c] + min(top_down(r-1,c,dp,grid), top_down(r,c-1,dp,grid))
                return dp[r][c]
        
        return top_down(len(grid)-1, len(grid[0])-1, dp, grid)

        

        
```
