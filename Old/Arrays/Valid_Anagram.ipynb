```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        """
        is t an anagram of s?

        s = a n a g r a m
        t = n a g a r a m

        s_dict = {
            a : 3
            n : 1
            g : 1
            r : 1
            m : 1
        }

        Hashmap/dict method
        """

        s_dict = defaultdict(int)
        t_dict = defaultdict(int)

        i = 0
        while i < min(len(s),len(t)):
            s_dict[s[i]]+=1
            t_dict[t[i]]+=1
            i+=1

        i = i-1
        
        if i < len(s)-1:
            while i < len(s):
                s_dict[s[i]]+=1
                i+=1
        elif i < len(t)-1:
            while i < len(t):
                t_dict[t[i]]+=1
                i+=1
        
        if s_dict==t_dict:
            return True
        return False



            

        
```
