```python
# NEED TO REVIEW
class Solution(object):
    def minHeightShelves(self, books, shelfWidth):
        """
        :type books: List[List[int]]
        :type shelfWidth: int
        :rtype: int
        """

        """
        DP
        What are we storing in the dp
        1. Bottom up
        What are things we need to build up
        What is the base case
        2. Top Down
        What is the sub-problem/recurrence relation for top down
        What is the base case

        What are the overlapping problems we can use memoization to avoid

        Storing in the dp: levels of books in a list that represent the min height

        https://leetcode.com/problems/filling-bookcase-shelves/solutions/2174413/python-sol-explained-with-picture-recursion-memo

        Comparing 2 options:
        total height as a result of putting on this level (if possible) or total height as a result of putting on next level
        Create a recursion tree (binary tree) of making these options
        
        Top Down
        At this level = min(total height as a result of putting the book on this level if possible, total height as a result of putting the book on the next level)
        """

        dp = {}
        def top_down(index,books,shelfWidth,width,height,dp):
            # WHY IS THIS MEMOIZED
            if (index,width,height) in dp:
                return dp[(index,width,height)]
            # Base Case
            if index == len(books)-1:
                return height
            
            if width + books[index+1][0] <= shelfWidth:
                # Choice 1: Put the next book on the same level
                # width argument: add the width of the next book
                # height argument: take the max current height of this level
                # returned height: maximum height of current level
                choice_1 = top_down(index+1,books,shelfWidth,width + books[index+1][0],max(height,books[index+1][1]),dp)

                # Choice 2: Put the next book on a different level
                # width argument: just the width of the next book (starting afresh on a new level)
                # height argument: just the height of the next book (starting afresh on a new level)
                # returned height: maximum height of current level + the height of the levels below it 
                choice_2 = height + top_down(index+1,books,shelfWidth,books[index+1][0],books[index+1][1],dp)

                # returned height:  minimum height as a result of making this choice
                dp[(index,width,height)] = min(choice_1,choice_2)
                return min(choice_1,choice_2)
            # Choice 2: Put the next book on a different level
            else:
                choice_2 = height + top_down(index+1,books,shelfWidth,books[index+1][0],books[index+1][1],dp)
                dp[(index,width,height)] = choice_2
                return choice_2

        return top_down(0,books,shelfWidth,books[0][0],books[0][1],dp)
```
