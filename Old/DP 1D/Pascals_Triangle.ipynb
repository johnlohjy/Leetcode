```python
"""
Need to optimise
"""

class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        
        """
        How to use dp to solve this problem
        Pascals triangle
        numrows = 1
        [[1]]
        Base case

        numrows = 2
        [[1], [1,1]]
        Base case

        numrows = 3
        [[1], [1,1], [1,2,1]]
        i=1

        numrows = 4
        [[1], [1,1], [1,2,1], [1,3,3,1]]
        i=1,i=2

        numrows = 5
        [[1], [1,1], [1,2,1], [1,3,3,1], [1,4,6,4,1]]
        i=1,i=2,i=3
        """

        dp = []
        for row_number in range(numRows):
            #Base Cases
            if row_number==0:
                dp.append([1])
            elif row_number==1:
                dp.append([1,1])
            else:
                #Initialise the numbers for that row
                temp = [1] * (row_number+1)
                #For the second to second last element, calculate its value
                for i in range(1,len(temp)-1):
                    temp[i] = dp[row_number-1][i-1] + dp[row_number-1][i]
                dp.append(temp)

        return dp
```
