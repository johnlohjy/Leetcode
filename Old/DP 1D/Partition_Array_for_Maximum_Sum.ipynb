```python
class Solution(object):
    def maxSumAfterPartitioning(self, arr, k):
        """
        :type arr: List[int]
        :type k: int
        :rtype: int
        """

        """
        [1,15,7,9,2,5,10] k=3

        Contigious arrays of AT MOST K

        [1,15,7] [9] [2,5,10]
        [15,15,15] [9] [10,10,10]


        [1] [15,7,9,2,5,10]
        Get the max sum for [1]: 1 and the result of the new sub problem [15,7,9,2,5,10]
            Get the max sum for [15]: 15 and the new sub problem [7,9,2,5,10]
            Get the max sum for [15,7]: 30 and the new sub problem [9,2,5,10]
            Get the max sum for [15,7,9]: 30 and the new sub problem [2,5,10]
            ^ Get the maximum result
        [1,15] [7,9,2,5,10]
        Get the max sum for [1,15]: 30 and the new sub problem [7,9,2,5,10]
        [1,15,7] [9,2,5,10]
        Get the max sum for [1,15,7]: 45 and the new sub problem [9,2,5,10]
        ^ Get the maximum result

        Memoization. Example [7,9,2,5,10]

        Store the result as dp[start index of partition]


        Recurrence relation:
        Store the result of starting the partitioning from i
        dp[i] = max(arr[:i+1])*(i+1) + top_down(arr[i+1:])

        dp[0] = max(arr[:1])*(i+1) + top_down(arr[1:])
        dp[0] = max(arr[:2])*(i+1) + top_down(arr[2:])
        dp[0] = max(arr[:3])*(i+1) + top_down(arr[3:])
        dp[0] = maximum possible value from partioning starting from i=0 to i=k-1

        return max dp

        ...

        DP
        1. Bottom up
        What are things we need to build up
        What is the base case
        2. Top Down
        What is the sub-problem/recurrence relation for top down
        What is the base case

        What are the overlapping problems we can use memoization to avoid

        Hint:
        Think dynamic programming: dp[i] will be the answer for array A[0], ..., A[i-1].
        For j = 1 .. k that keeps everything in bounds, dp[i] is the maximum of dp[i-j] + max(A[i-1], ..., A[i-j]) * j .
        """
        dp = {}
        def top_down(start_i, arr, k, dp):
            if start_i in dp:
                return dp[start_i]
            dp[start_i] = 0
            for i in range(start_i,start_i+k):
                if i==len(arr):
                    break
                temp = max(arr[start_i:i+1])*(len(arr[start_i:i+1])) + top_down(i+1,arr,k,dp)
                if temp > dp[start_i]:
                    dp[start_i] = temp
            return dp[start_i]
        
        top_down(0,arr,k,dp)
        return max(dp.values())



        
```
