```python
class Solution(object):
    def findLongestChain(self, pairs):
        """
        :type pairs: List[List[int]]
        :rtype: int
        """

        """
        n pairs
        pairs[i] = [left, right]. left < right
        p1 = [a,b]
        p2 = [c,d]
        if b < c, a chain can be formed

        Return the longest chain which can be formed

        DP
        1. Bottom up
        What are things we need to build up
        What is the base case
        2. Top Down
        What is the sub-problem/recurrence relation for top down
        What is the base case

        What are the overlapping problems we can use memoization to avoid

        [[1,2], [2,3], [3,4]]
        [1,2] -> [3,4]

        ##############################################################################

        [[1,2], [7,8], [4,5]]
        [1,2] -> [4,5] -> [7,8]
        Ans(2):
        [1,2] -> [7,8]

        top down dp:
        [1,2] -> [4,5]
        [1,2] -> [4,5] -> [7,8]

        [7,8]

        [4,5] -> [7,8]

        Overlapping problems:
        [4,5] -> [7,8]

        For each pair, build its chain
        What is the longest chain for [1,2]
        => what is the longest chain for [7,8]
        => what is the longest chain for [4,5]
           => what is the longest chain for [7,8]

        What is the longest chain for [7,8]
        => Cannot check [1,2]
        => Cannot check [4,5]

        What is the longest chain for [4,5]
        => Cannot check [1,2]
        => what is the longest chain for [7,8]

        dp[0] = max(1+top_down(dp[1]),1+top_down(dp[2]))
        dp[1] = Base Case return 1
        dp[2] = max(1+top_down(dp[1]))
        """
        dp = {}
        def top_down(i,pairs):
            if i in dp:
                return dp[i]
            else:
                curr_val = 1
                for j in range(len(pairs)):
                    if i==j:
                        continue
                    if pairs[i][1] < pairs[j][0]:
                        temp = 1 + top_down(j,pairs)
                        if temp > curr_val:
                            curr_val = temp
                dp[i] = curr_val
                return dp[i]

        for i in range(len(pairs)):
            top_down(i,pairs)
        return max(dp.values())
                    


        
```
