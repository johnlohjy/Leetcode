```python
class Solution(object):
    def numSquares(self, n):
        """
        :type n: int
        :rtype: int
        """

        """
        Given n,
        return the least number of perfect square numbers that sum to n

        Perfect square number: Can be square root to int
        Index:  1,2,3,4,5,6,7,8
        Square: 1,4,9,16,25,36,49,64

        DP
        1. Bottom up
        What are things we need to build up
        What is the base case
        2. Top Down
        What is the sub-problem/recurrence relation for top down
        What is the base case

        What are the overlapping problems we can use memoization to avoid

        Example
        Number 12

        Get the list of square numbers less than 12
        Candidates: 1,4,9

        Top Down DP
        dp[n]: Memoize least number of perfect square numbers that sum to n

        dp[12] = min(dp[3],dp[8],dp[11])
                      -9    -4     -1

        

        Base Case: Not a perfect square number
        """

        squares = []
        for i in range(1,n+1):
            squares.append(i**2)

        dp = {}

        def top_down(current,dp,squares):
            # Memoization
            if current in dp:
                return dp[current]

            # Base Cases
            # Base Case 1: No more things to add
            if current == 0:
                return 0

            if current < 0:
                return 10000
            
            # Recursive Case
            # least is capped at 10000 (+1)
            least = 10000
            for i in squares:
                if i > current:
                    break
                temp = 1 + top_down(current-i,dp,squares)
                if temp < least:
                    least = temp
            
            dp[current] = least
            return dp[current]


        return top_down(n,dp,squares)
        
```
