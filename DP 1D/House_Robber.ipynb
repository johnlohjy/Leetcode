```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        
        """
        Cannot rob adjacent houses because alarm will ring

        [1,2,3,1]
        1+3 = 4

        [2,7,9,3,1]
        2+9+1 = 12
        
        [20,1,1,23,1,5,6]
        20+23+6
        Max possible money we can rob so far as at each time step 
        - when adding, we cannot add to adjacent
        [20,20,21,43,43,48,49]
        Note: At the second house we put 20.
        For example when we are at the 4th house, we can rob from the 3rd, 2nd or 1st house
        but because we don't want to look back so much, we look back to the 2nd house only as it represents
        the max loot we could have gotten so far. At each house, we take the max of:
        - Looting the current house and adding it to the cumulated loot of the prev prev house
        - Taking the cumulated loot of the prev house 
        Therefore each house now represents the max loot we could have gotten so far

        How to use dynamic programming to find the max amount of money
        we can rob?
        """
        return self.bottom_up_dp(nums)

    def bottom_up_dp(self,nums):
        memo = {}
        #Base Cases
        memo[-2]=0
        memo[-1]=0
        for i in range(len(nums)):
            #At each house,
            #Take the max of
            #Looting the current house and adding it to the cumulated loot of the prev prev house
            #Taking the cumulated loot of the prev house 
            memo[i] = max(nums[i] + memo[i-2], memo[i-1])

        #Return the max loot possible
        return memo[len(nums)-1]
        


```
