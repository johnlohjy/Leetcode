```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    res = 0
    def maxAncestorDiff(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        """
        For each subtree, return the minimum value and maximum value of its descendants and find the absolute difference

        1) Post order traversal:
        Left-Right-Root Policy
        The left subtree is traversed first
        Then the right subtree is traversed
        Finally, the root node of the subtree is traversed
            

        pseudocode:
        def postorder(node):
            if node == None:
                return
            postorder(node.left)
            postorder(node.right)
            print(node.val)
        """
        def dfs(node):
            # If the node's child is none, return an empty list
            if node==None:
                return []

            # Initialise the node's left and right subtree max and min descendants
            l_res = []
            r_res = []

            # If the left and right subtrees max and min descendants exist, extend the result
            if node.left:
                l_res.extend(dfs(node.left))
            if node.right:
                r_res.extend(dfs(node.right))
            # Combine the max and min descendants 
            temp_desc = l_res + r_res

            # If there were no max and min descendants, return its own value
            if temp_desc==[]:
                return [node.val]
            
            # If there were max and min descendants, get the max abs diff between this node and its max and min descendants
            # Store the max result so far
            temp_diff = max( abs(node.val-max(temp_desc)), abs(node.val-min(temp_desc)) )
            if temp_diff > self.res:
                self.res=temp_diff
            
            # Return the node's own value, and the max and min descendants
            return [node.val,max(temp_desc),min(temp_desc)]

        dfs(root)
        return self.res
```
