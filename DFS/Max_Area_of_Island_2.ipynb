```python
class Solution(object):
    def maxAreaOfIsland(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        
        """
        Given m x n grid

        Island: Group of 1s (representing land) connected 4-directionally

        Area of an island is the number of cells with a value of 1 in the island

        Return the max area of the island in the grid
        """

        visited = set()

        def dfs(r,c,grid,visited):
            # Check if the current cell has been visited before
            if (r,c) not in visited:

                visited.add((r,c))

                # Check if the current cell is out of bounds or a 0
                # If either of the above conditions are T, return 0 to add on to the area
                if r < 0 or r >= len(grid) or c < 0 or c >= len(grid[0]) or grid[r][c]==0:
                    return 0

                # If this is a valid cell that is part of the island
                # return its own area + the area of other possible islands searched dfs
                return 1 + dfs(r+1,c,grid,visited) + dfs(r-1,c,grid,visited) + dfs(r,c-1,grid,visited) + dfs(r,c+1,grid,visited)
            return 0

        # Initialise an empty areas list
        areas = []

        # Iterate through the grid to perform DFS
        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if (r,c) not in visited:
                    areas.append(dfs(r,c,grid,visited))

        return max(areas)

                




```
