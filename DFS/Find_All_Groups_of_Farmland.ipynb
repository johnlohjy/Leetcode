```python
class Solution(object):
    def findFarmland(self, land):
        """
        :type land: List[List[int]]
        :rtype: List[List[int]]
        """

        """
        land: m x n binary matrix
        0: forested land
        1: farmland

        Groups: rectangular areas of farmland
        No 2 groups are adjacent (four directionally)

        Find the coordinates of the top left and bottom right
        of each group of farmland:
        [r1,c1 , r2,c2]
        [top left, bottom right]

        Return 2D array containing the 4-length arrays
        return in any order 

        Hint:
        Since every group of farmland is rectangular, the top left corner of each group will have the smallest x-coordinate and y-coordinate of any farmland in the group.

        Similarly, the bottom right corner of each group will have the largest x-coordinate and y-coordinate of any farmland in the group.

        Use DFS to traverse through different groups of farmlands and keep track of the smallest and largest x-coordinate and y-coordinates you have seen in each group.
        """

        visited = set()

        def dfs(r,c,visited,land,farmland):
            if (r,c) not in visited:
                visited.add((r,c))

                if r < 0 or r > len(land)-1 or c < 0 or c > len(land[0])-1:
                    return

                if land[r][c]==1:
                    farmland.append((r,c))
                    dfs(r+1,c,visited,land,farmland)
                    dfs(r-1,c,visited,land,farmland)
                    dfs(r,c-1,visited,land,farmland)
                    dfs(r,c+1,visited,land,farmland)
                    return
                return
            return

        def farmland_coord(farmland,res):
            if len(farmland)==1:
                res.append([farmland[0][0], farmland[0][1], farmland[0][0], farmland[0][1]])
                return

            min_coord = [farmland[0][0], farmland[0][1]]
            max_coord = [farmland[0][0], farmland[0][1]]

            for i in range(1,len(farmland)):
                temp = farmland[i][0] + farmland[i][1]
                if temp < min_coord[0]+min_coord[1]:
                    min_coord = [farmland[i][0],farmland[i][1]]
                
                if temp > max_coord[0]+max_coord[1]:
                    max_coord = [farmland[i][0],farmland[i][1]]
            
            res.append(min_coord+max_coord)
            return

        res = []
        for r in range(len(land)):
            for c in range(len(land[0])):
                if (r,c) not in visited:
                    farmland = []
                    dfs(r,c,visited,land,farmland)
                    if len(farmland) > 0:
                        farmland_coord(farmland,res)
        
        return res

        
                     



        
```
