```python
class Solution(object):
    def countSubIslands(self, grid1, grid2):
        """
        :type grid1: List[List[int]]
        :type grid2: List[List[int]]
        :rtype: int
        """

        """
        2 m x n binary matrics grid1 and grid2
        - 0s: water
        - 1s: land

        Island: Group of 1s connected 4 directionally
        
        An island in grid2 is considered a sub-island 
        if there is an island in grid1 that contains all the cells that make up this island in grid2

        Return the number of islands in grid2 that are considered sub-islands

        Approach:
        - Get all islands in grid 2
        - For each island, check if all the cells are 1s in grid 1. If they are, then they are sub-islands 
        """

        visited = set()

        def dfs(r,c,grid,visited,islands):
            if (r,c) not in visited:
                visited.add((r,c))

                if r < 0 or r >= len(grid) or c < 0 or c >= len(grid[0]) or grid[r][c]==0:
                    return

                islands.append((r,c))

                dfs(r-1,c,grid,visited,islands)
                dfs(r+1,c,grid,visited,islands)
                dfs(r,c-1,grid,visited,islands)
                dfs(r,c+1,grid,visited,islands)

                return
            return
        
        def sub_island(grid_1,islands):
            for r,c in islands:
                if grid_1[r][c] == 0:
                    return False
            return True

        sub_islands = 0

        for r in range(len(grid2)):
            for c in range(len(grid2[0])):
                if (r,c) not in visited:
                    islands = []
                    dfs(r,c,grid2,visited,islands)
                    # print(islands)
                    # print("")
                    if len(islands) >0 and sub_island(grid1,islands):
                        sub_islands+=1

        return sub_islands







        
```
