```python
"""
Unoptimised
"""

class Solution(object):
    def countBattleships(self, board):
        """
        :type board: List[List[str]]
        :rtype: int
        """

        """
        Battleship shape: 
        1 x k. 1 row k col
        k x 1. k col 1 row

        At least 1 horizontal or vertical cell separates the 2 battleships 

        X => battleship
        . => empty

        Return the number of battleships

        DFS problem
        For each cell 
            1. Recursively look left/right
            2. Recurisvely look up/down
        Stopping condition: 
            1. Out of bounds
            2. Empty cell
        """
        
        visited = set()

        def dfs(r,c,visited,board):
            if (r,c) not in visited:
                visited.add((r,c))

                if r < 0 or r > len(board)-1 or c < 0 or c > len(board[0])-1 or board[r][c]=='.':
                    return 0
                else:
                    return 1 + dfs(r+1, c, visited, board) + dfs(r-1, c, visited, board) + dfs(r, c-1, visited, board) + dfs(r, c+1, visited, board)

            return 0

        res = 0

        for r in range(len(board)):
            for c in range(len(board[0])):
                search = dfs(r,c,visited,board)
                if search > 0:
                    res+=1

        return res
```
