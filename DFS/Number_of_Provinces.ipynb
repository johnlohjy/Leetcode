```python
class Solution(object):
    def findCircleNum(self, isConnected):
        """
        :type isConnected: List[List[int]]
        :rtype: int
        """

        """
        A group of 'connected' cities = 1 province

        For each city, perform DFS on its connections to find a province. 

        Use visited to keep track of cities which we have performed DFS on

        Use province to keep track of the cities belonging to the same province (same DFS)
        """

        visited = set()
        provinces = []

        #Perform a province DFS on each city
        # - some cities may not be a part of a province (disconnected)
        for j in range(len(isConnected)):
            province = set()
            self.dfs(isConnected,visited,j, province)
            #If there is a province present after the DFS, then add it to our list of provinces
            if(len(province)>0):
                provinces.append(province)
        #return the number of provinces
        return len(provinces)

        
    def dfs(self, isConnected, visited, cityIndex, province):
        #If the city has not been visited before
        if cityIndex not in visited:
            #Add the city to visited 
            #Its only visited if
            # - We start our search from here at the very beginning
            # - It has been part of a province dfs 
            visited.add(cityIndex)
            #Add the city to the province dfs. Different from visited because visited
            #contains all visited nodes, where as province only contain the nodes 
            #that are a part of the same province
            province.add(cityIndex)

            #Peform DFS on the city's connections
            #For each connection
            for i in range(len(isConnected[cityIndex])):
                #If the city has not been visited before 
                if i not in visited:
                    #If there is a connection
                    if isConnected[cityIndex][i]==1:
                        #Continue the province DFS from that city onwards
                        self.dfs(isConnected,visited,i,province)
```
