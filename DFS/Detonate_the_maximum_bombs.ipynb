```python
class Solution(object):
    def maximumDetonation(self, bombs):
        """
        :type bombs: List[List[int]]
        :rtype: int
        """
        
        """
        Range of bomb: Area where its effects can be felt (Shape of a circle with the centre as the location of the bomb)

        bombs[i] = [x-coord,y-coord,radius]

        You can detonate a single bomb
        When a bomb is detonated, it will detonate all bombs that lie in its range 
        These bombs will further detonate the bombs that lie in their ranges

        Given the list of bombs, return the max number of bombs that can be detonated if you are allowed to detonate only 1 bomb

        Hints:
        - How can we model the relationship between different bombs? Can "graphs" help us?
        - Bombs are nodes and are connected to other bombs in their range by directed edges.
        - If we know which bombs will be affected when any bomb is detonated, how can we find the total number of bombs that will be detonated if we start from a fixed bomb?
        - Run a Depth First Search (DFS) from every node, and all the nodes it reaches are the bombs that will be detonated.
        """

        def can_detonate(bomb1,bomb2):
            distance = ((bomb1[0]-bomb2[0])**2 + (bomb1[1]-bomb2[1])**2)**0.5
            if distance <= bomb1[2]:
                return True
            return False

        adj_list = defaultdict(list)
        for i in range(len(bombs)):
            for j in range(len(bombs)):
                if i==j:
                    continue
                if can_detonate(bombs[i],bombs[j]):
                    adj_list[i].append(j)

        def dfs(node,visited):
            if node not in visited:
                visited.add(node)
                res = 1
                for nb in adj_list[node]:
                    res += dfs(nb,visited)
                return res
            return 0

        res = 0

        for i in range(len(bombs)):
            visited = set()
            temp = dfs(i,visited)
            if temp > res:
                res = temp
        
        return res


        
        
```
