```python
class Solution:
    def champagneTower(self, poured: int, query_row: int, query_glass: int) -> float:
        """
        :type poured: int
        :type query_row: int
        :type query_glass: int
        :rtype: float
        """

        """
        1 cup of champagne:
        - top glass is full

        2 cups of champagne:
        - top glass is full
        - second glasses are half full

        3 cups of cups of champagne
        - top glass is full
        - second glasses are full

        4 cups of champagne
        - top glass is full: 1. Excess is 3. Excess is 3
        - second glasses are full: Each glass get 3/2=1.5. Excess is 0.5
        - outer glasses are quarter full: Each glass get 0.5/2=1/4
        - middle glass is half full: 1/4+1/4=0.5

        First row has 1 glass
        Second row has 2 glasses
        Third row has 3 glasses 
        etc.

        Return how full the jth glass in the ith row is

        DP
        1. Bottom up
        What are things we need to build up
        What is the base case
        2. Top Down
        What is the sub-problem/recurrence relation for top down
        What is the base case

        What are the overlapping problems we can use memoization to avoid 

        What to store
        1 cup
        dp[0][0] = 1 cup

        2 cups
        dp[0][0] = 1 cup
        dp[1][0] = 1 cup
        dp[1][1] = 1 cup

        3 cups
        dp[0][0] = 1 cup
        dp[1][0] = 1 cup
        dp[1][1] = 1 cup
        dp[2][0] = 1/4 cup
        dp[2][1] = 1/2 cup
        dp[2][2] = 1/4 cup

        [[0],[0,0],[0,0,0]]

        top_down(dp[0][0]) = top_down(dp[1][0]), top_down(dp[1][1])

        top_down(dp[i][j]) = top_down(dp[i+1][j from 0 to nth row])

        Base Case: no more

        Apply memoization:
        Row: 2
        - 0 and 1 is the same

        Row: 3
        - 0 and 2 is the same

        Row: 4
        - 
        """

        # Make dp
        dp = []
        for i in range(1,101):
            dp.append([0]*i)
        
        def top_down(i,j,liquid,dp):
            #Base Case
            if liquid==0:
                return

            #Recursive Case
            #If curr glass + liquid <= 1 glass, there won't be excess. Keep it as it is
            if dp[i][j] + liquid <= 1:
                dp[i][j] += liquid
                return
            #Else, there will be excess
            else:
                #Calculate the excess
                #If it currently has liquid
                if dp[i][j] > 0:
                    excess = (dp[i][j] + liquid) -1
                #If it doesnt have liquid
                else:
                    excess = liquid-1
                #Assign 1 to the glass since it will have excess
                dp[i][j] = 1
                top_down(i+1,j,excess/2,dp)
                top_down(i+1,j+1,excess/2,dp)

        top_down(0,0,poured,dp)

        return dp[query_row][query_glass]
```
